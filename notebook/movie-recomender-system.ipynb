```python
# columns that are necessary 
# genere , id , keywords , title , overview , cast , crew

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
# check for any null value
movies.isnull().sum()
```

```python
# Since overview is an important colummn the null values should be droped
movies.dropna(inplace=True)
```

```python
# check whether there is any duplicate data or not
movies.duplicated().sum()
```

```python
# create a funtion to make the column more accurate
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj): # first we have to convert the strings into a list hats why we use ast library
        L.append(i['name'])
    return L
```

```python
# In genre column we need the tags like action , sci-fi, thrillar etc.... they are in a list with the title name 
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
# Same goes for keywords
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# But for the cast there is to much name thats why we need only top 3 cast 
# Create another funtion to extract top 3 cast from list
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3 :
            L.append(i['name'])
            counter+=1
        else :
            break
    return L
```

```python
 movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
# In crew we need only director name 
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break 
    return L
```

```python
  movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# Overview column is a string we need to convert it into a list
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# Removing blank space and make a single token
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# Making a single colmn named as tags
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
# Steaming
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse = True , key = lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman')
```
